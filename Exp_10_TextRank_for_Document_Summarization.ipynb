```python
import numpy as np
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer

text = """Nature is calm and beautiful, offering peace and freshness to our lives. The green trees, fresh air, and gentle breeze help us relax and feel happy. It reminds us to enjoy the simple and pure moments around us."""

sentences = text.split(". ")
sentences = [s.strip() for s in sentences if s]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
similarity_matrix = (X * X.T).toarray()

# Step 5: Build graph
nx_graph = nx.from_numpy_array(similarity_matrix)
scores = nx.pagerank(nx_graph)
ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)
print("Summary:")
for score, sentence in ranked_sentences[:2]:
    print("-", sentence)
```
