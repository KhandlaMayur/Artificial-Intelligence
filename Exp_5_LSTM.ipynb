```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Step 1: Create dataset (simple sequence)
def create_dataset():
    data = np.array([i for i in range(10)])  # 0 to 9
    X, y = [], []

    for i in range(len(data) - 3):
        X.append(data[i:i+3])   # 3 input steps
        y.append(data[i+3])     # next value

    return np.array(X), np.array(y)

X, y = create_dataset()

# Reshape input to [samples, time steps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))

# Step 2: Build LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(3, 1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')

# Step 3: Train model
model.fit(X, y, epochs=200, verbose=0)

# Step 4: Test prediction
test_input = np.array([7, 8, 9])
test_input = test_input.reshape((1, 3, 1))

prediction = model.predict(test_input)
print("Predicted next value:", prediction[0][0])
```
